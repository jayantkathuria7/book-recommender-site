# Popularity-based Recommender System

```python
ratings_with_names = ratings.merge(books, on='ISBN')
ratings_with_names
```

```python
num_rating_df = ratings_with_names.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_names.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
avg_rating_df.sort_values(by='avg_ratings', ascending=False)[140:160]
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
```

```python
popular_df
```

```python
popular_df =  popular_df[popular_df['num_ratings']>=250].sort_values(by='avg_ratings', ascending = False)[:50]
popular_df
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'avg_ratings', 'num_ratings', 'Image-URL-M']]
```

```python
popular_df.shape
```

# Collaborative Filtering Based Recommender system

```python
x = ratings_with_names.groupby('User-ID').count()['Book-Rating']>200
good_users = x[x].index
good_users
```

```python
filtered_rating = ratings_with_names[ratings_with_names['User-ID'].isin(good_users)]
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
popular_books = y[y].index
popular_books
```

```python
final_df = filtered_rating[filtered_rating['Book-Title'].isin(popular_books)]
final_df.head()
```

```python
final_df.shape
```

```python
pt = final_df.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
pt.fillna(0, inplace = True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score
```

```python
def recommend(book_name):
    index =  np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```
